```python
import pandas as pd
df = pd.read_csv('synthetic_efdata.csv')
df.head()
```

```python
#data cleaning
#missing values 
df.fillna(df.mean(), inplace =True)

#duplicates
df.drop_duplicates(inplace = True)

#check
print("Duplicate Rows:", df.duplicated().sum())
print("Missing Values:", df.isnull().sum())
```

```python
#data type conversion
df['Income ($)'] = df['Income ($)'].astype(int)
df['Monthly Expenses ($)'] = df['Monthly Expenses ($)'].astype(int)
df['Dependents'] = df['Dependents'].astype(int)
df['Job Stability (1-10)'] = df['Job Stability (1-10)'].astype(int)
df['Lifestyle Cost (1-10)'] = df['Lifestyle Cost (1-10)'].astype(int)
df['Recommended Emergency Fund ($)'] = df['Recommended Emergency Fund ($)'].astype(int)

#outly
import numpy as np
from scipy import stats

z_scores = np.abs(stats.zscore(df.select_dtypes(include=np.number))) #only select nums
df = df[(z_scores < 3).all(axis=1)] #threshold 3sd away
```

```python
df
```

```python
#standardizing
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Income ($)', 'Monthly Expenses ($)', 'Dependents', 'Job Stability (1-10)', 'Lifestyle Cost (1-10)']] = scaler.fit_transform(df[['Income ($)', 'Monthly Expenses ($)', 'Dependents', 'Job Stability (1-10)', 'Lifestyle Cost (1-10)']])
```

```python
#save
df.to_csv('cleaned_efData.csv', index=False)
```
